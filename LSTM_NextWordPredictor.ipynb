```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faq])
```

```python
tokenizer.word_counts
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faq.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_length =max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input_sequences, maxlen  =max_length, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes = 539)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(539, 100, input_shape = (83,)))
model.add(LSTM(150))
model.add(Dense(539, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y, epochs = 100)
```

```python
text = "robot"
import numpy as np
for i in range(10):
  token_text = tokenizer.texts_to_sequences([text])[0]
  padded_token_text = pad_sequences([token_text], maxlen = max_length, padding='pre')
  pred = model.predict(padded_token_text)

  pos = np.argmax(pred, axis=1)[0]

  output_word = tokenizer.index_word.get(pos,"")
  text = text+ " "+ output_word

  print(text)
```
